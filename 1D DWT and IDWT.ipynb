```python
%matplotlib inline
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
import pywt
```

# Generating noisy signal

```python
freq = 50
time_period = 1/freq #20ms
time = time_period * 2 #40ms
amplitude = 2
```

```python
t = np.linspace(0, time, 500, endpoint = True)
x = 2*3.14*freq*t 
```

```python
yc = amplitude * np.sin(x) + np.random.randn(len(x))
```

```python
fig1 = plt.figure(figsize=(10, 4))
axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,yc)
```

# Discret Wavelet Transformation of signal 

```python
ca, cd =pywt.dwt(yc,'db5')
ca2, cd2 = pywt.dwt(ca,'db5')
ca3, cd3 = pywt.dwt(ca2, 'db5')
```

```python
print len(yc)    
print len(ca)
print len(ca2)
print len(ca3)
```

```python
plt.plot(yc)
plt.plot(ca)
plt.plot(ca2)
plt.plot(ca3)
```

## Inverse Discret Wavelet Transformation of denoised signal

```python
A = pywt.idwt(ca, None, 'db5')
A2 = pywt.idwt(ca2, None, 'db5')
A3 = pywt.idwt(ca3, None, 'db5')
```

```python
plt.plot(yc)
plt.plot(A)
plt.plot(A2)
plt.plot(A3)
```

```python
A2_prime = pywt.idwt(A2, None, 'db5')
A3_prime = pywt.idwt(A3, None, 'db5')
```

```python
plt.plot(yc)
plt.plot(A)
plt.plot(A2_prime)
plt.plot(A3_prime)
```

```python
A3_secundo = pywt.idwt(A3_prime, None, 'db1')
```

```python
plt.plot(yc)
plt.plot(A)
plt.plot(A2_prime)
plt.plot(A3_secundo)
```
