## Exploring certain function

```python
rbio_5_5 = pywt.Wavelet('rbio5.5')
```

```python
print rbio_5_5
```

```python
phi_d, psi_d, phi_r, psi_r, x = rbio_5_5.wavefun(level=10)
```

```python
f, axarr = plt.subplots(2, sharex=True)
f.set_size_inches(10,10)
axarr[0].set_title('Reconstructing functions', fontsize=14, fontweight='bold')
axarr[0].plot(x, phi_r, color = 'red', label = 'Scalling')
axarr[0].plot(x, psi_r, color = 'green', label = 'Wavelet')
axarr[0].legend()
axarr[1].set_title('Decompositions functions', fontsize=14, fontweight='bold')
axarr[1].plot(x, phi_d, color = 'red', label = 'Scalling')
axarr[1].plot(x, psi_d, color = 'green', label = 'Wavelet')
axarr[1].legend()
```

## Custom wavelet

```python
r = math.sqrt(2)/3
d = math.sqrt(5)/2
dec_lo, dec_hi, rec_lo, rec_hi = [r, r/3], [-d, d/3], [r, r/2], [d, -d/2]
filter_bank = [dec_lo, dec_hi, rec_lo, rec_hi]
myWavelet = pywt.Wavelet(name="myHaarWavelet", filter_bank=filter_bank)
```

```python
phi_d, psi_d, phi_r, psi_r, x = myWavelet.wavefun(level=10)
```

```python
f, axarr = plt.subplots(2, sharex=True)
f.set_size_inches(10,10)
axarr[0].set_title('Reconstructing functions', fontsize=14, fontweight='bold')
axarr[0].plot(x, phi_r, color = 'red', label = 'Scalling')
axarr[0].plot(x, psi_r, color = 'green', label = 'Wavelet')
axarr[0].legend()
axarr[1].set_title('Decompositions functions', fontsize=14, fontweight='bold')
axarr[1].plot(x, phi_d, color = 'red', label = 'Scalling')
axarr[1].plot(x, psi_d, color = 'green', label = 'Wavelet')
axarr[1].legend()
```
